Mars Hemispheres

```python
#create and executable path and browser to scrape for images
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

#link the URL for scraping
url = 'https://marshemispheres.com/'
browser.visit(url)

#create an html object
html = browser.html
#Create the BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
#collect headings assiciated with the images before grabbing images, so we know which images we are looking for. 
mars_headers=[]
titles = soup.find_all('h3')
titles
```

```python
#removing the last header 'Back' that we don't need
titles = titles[:-1]
```

```python
#with a loop, get rid of the html tags to leave us with just the header text
for title in titles:
    mars_headers.append(title.text)
    
mars_headers
```

```python
#now put the four titles and images together
mars_titles_images=[]
count=0

for i in mars_headers:
    #using splinter to find thumbnail images, click through to each link,
    #then find the src link for each image with class "wide-image"
    browser.find_by_css('img.thumb')[count].click()
    mars_titles_images.append(browser.find_by_css('img.wide-image')['src'])
    browser.back()
    #counter used to loop through each thumbnail image
    count=count+1

mars_titles_images
```

```python
#list to store dictionaries in
hemisphere_image_urls = []
counter = 0
for i in mars_titles_images:
    #appending each title and url to the list of dictionaries
    hemisphere_image_urls.append({"title":mars_headers[counter],"img_url":mars_titles_images[counter]})
    counter = counter+1
```
